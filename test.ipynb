```python
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.document_loaders import TextLoader
```

```python
_FILE_PATH = "data/sample_docs.txt"



loader = TextLoader(_FILE_PATH, encoding="utf-8", autodetect_encoding=True)
docs = loader.load()
```

```python
text_splitter = CharacterTextSplitter(
    chunk_size=100,    # 每个块的目标大小为100个字符
    chunk_overlap=10   # 每个块之间重叠10个字符，以缓解语义割裂
)
```

```python
chunks = text_splitter.split_documents(docs)
print(len(chunks))
```

```python
print(f"文本被切分为 {len(chunks)} 个块。\n")
print("--- 前5个块内容示例 ---")
for i, chunk in enumerate(chunks[:5]):
    print("=" * 20)
    print(f'切片 {i+1} (长度:{len(chunk.page_content)}):" {chunk.page_content}"')
```
